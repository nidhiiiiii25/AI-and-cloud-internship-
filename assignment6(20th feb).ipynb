```python
import time
import numpy as np

# ---------- Python List Test ----------

start = time.time()

python_list = list(range(1000000))
python_list = [x * 2 for x in python_list]

end = time.time()

print("Python List Time:", end - start)


# ---------- NumPy Array Test ----------

start = time.time()

numpy_array = np.arange(1000000)
numpy_array = numpy_array * 2

end = time.time()

print("NumPy Array Time:", end - start)
```

Observation 1
NumPy arrays execute numerical operations significantly faster than Python lists when working with large datasets.


observation 2
Python lists require explicit looping to perform operations, whereas NumPy performs vectorized operations that process all elements simultaneously.


Observation 3
NumPy is more efficient for scientific computing and data analysis tasks involving large numerical data.
