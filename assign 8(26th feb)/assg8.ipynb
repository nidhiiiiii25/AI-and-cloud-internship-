```python
import pandas as pd
```

```python
df=pd.read_csv("assg8.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["city"]=df["city"].fillna("unknown")
```

```python
df.isnull().sum()
```

```python
df["rating"].unique()
```

```python
df.groupby("city")["rating"].transform("median")
```

```python
df["rating"] = df["rating"].fillna(
    df.groupby("city")["rating"].transform("median")
    
)

df["rating"] = df["rating"].fillna(df["rating"].median())
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
# Fix case issues
df["city"] = df["city"].str.lower()

# Remove spaces
df["city"] = df["city"].str.strip()
```

“Data cleaning ensures accuracy, consistency, and reliability of analysis by handling missing values, removing duplicates, and standardizing data.”
