Insights

1. Web Development Masterclass has the highest number of paid students.
This indicates that web development skills are highly demanded among learners on the platform.

2. Some courses have more free students than paid students.
Courses such as Photography Basics and Digital Marketing 101 attract many free learners, suggesting that introductory or hobby-based courses often have higher free enrollments.

3. The IT category appears multiple times in the dataset.
This suggests that technology-related courses are a major focus area on the platform.

4. Course durations vary widely.
The dataset contains courses ranging from short courses (around 6 hours) to long courses (up to 30 hours), indicating both beginner-friendly and in-depth programs.

5. The dataset contains missing data in the free_students column.
This highlights the importance of handling missing values before performing deeper analysis.
