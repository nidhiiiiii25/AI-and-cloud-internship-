```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Q1: Which course category has the highest number of students?
# Q2: What is the proportion of students who completed vs in progress?
# Q3: How are the course ratings distributed?

```python
df=pd.read_csv('data.csv')
```

```python
# Q1: Which course category has the highest number of students?
category_count=df['category'].value_counts()
category_count.plot(kind='bar')
plt.xlabel("Category")
plt.ylabel("Number of Students")
plt.title("Student Distribution by Category")
plt.show()
```

```python
# Q2: What is the proportion of students who completed vs in progress?
status_count = df['completion_status'].value_counts()
status_count.plot(kind='pie', autopct='%1.1f%%')
plt.title("Completion Status Distribution")
plt.ylabel("")  # removes extra label
plt.show()
```

```python
# Q3: How are the course ratings distributed?
df['rating'].plot(kind='hist')
plt.title("Rating Distribution")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.show()
```

The analysis of student course data reveals that the IT category is the most popular among students, indicating a strong interest in technical skills such as programming and web development.

A significant proportion of students have successfully completed their courses, suggesting good engagement and commitment towards learning. However, a smaller group is still in progress, which may indicate ongoing learning or course difficulty.

Additionally, the course ratings are generally high, with most ratings clustered above 4.0. This suggests that students are largely satisfied with the courses offered.

Overall, the data indicates that IT courses are in high demand, have good completion rates, and receive positive feedback from learners
